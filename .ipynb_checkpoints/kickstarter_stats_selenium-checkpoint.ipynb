```python
from bs4 import BeautifulSoup
import requests
import time, os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
url = 'https://www.kickstarter.com/help/stats'
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
driver = webdriver.Chrome(chromedriver)
driver.get(url)
```

```python
# Access the first table's contents
table = driver.find_element_by_class_name('table_container')
table.text
```

```python
# Click the "See categories" button to expand the table elements
driver.find_element_by_class_name('details').click()
```

```python
table = driver.find_element_by_class_name('table_container')
table.text
```

```python
# Convert text to list
rows = table.text.split('\n')
rows
```

```python
# Split headers and data; clean the data
headers = ['Category', 'Launched Projects', 'Total Dollars', 'Successful Dollars', 'Unsuccessful Dollars', 'Live Dollars', 'Live Projects', 'Success Rate']
data = rows[23:-1]
data
```

```python
# Remove non-numeric characters in the data
clean_rows = [row.replace('$', '').replace('B','').replace('M', '').replace('K', '').replace('%', '').replace(',', '') for row in data]
clean_rows   
```

```python
# Create list of lists splitting by spaces
split_data = [row.split(' ') for row in clean_rows]
split_data
```

```python
# Remove empty strings
for data in split_data:
    while ("" in data):
        data.remove("")

split_data
```

```python
# Clean misformatted columns by spot checking
# Clean 'Film & Video' category name
split_data[3] = split_data[3][2:]
split_data[3][0] = 'Film & Video'

# Clean 'Music' category name
split_data[4][0] = 'Music'

split_data
```

```python
# Convert numeric strings to floats
kickstarter_stats = []
for data in split_data:
    rows = []
    for string in data:
        try:
            rows.append(float(string))
        except:
            rows.append(string)

    kickstarter_stats.append(rows)

kickstarter_stats
```

```python
# Create a dataframe using the data and column headers
df = pd.DataFrame(kickstarter_stats, columns=headers)
df
```

```python
# Clean amounts to eliminate abbreviations in original table ('B', 'M', 'K')
# Convert 'Total Dollars'
df['Total Dollars'] = df['Total Dollars'] * 1000000
df.loc[:1, 'Total Dollars'] *= 1000

# Convert 'Successful Dollars'
df['Successful Dollars'] = df['Successful Dollars'] * 1000000
df['Successful Dollars'][0] = df['Successful Dollars'][0] * 1000

# Convert 'Unsuccessful Dollars'
df['Unsuccessful Dollars'] = df['Unsuccessful Dollars'] * 1000000

# Convert 'Live Dollars'
for idx, row in df.iterrows():
    if df.loc[idx,'Live Dollars'] < 1000:
        df.loc[idx,'Live Dollars'] *= 1000
df.loc[:2, 'Live Dollars'] *= 1000

# Convert 'Success Rate' (to a percentage)
df['Success Rate'] /= 100
```

```python
# Set the categories as the dataframe index
df.set_index('Category', inplace=True)
```
